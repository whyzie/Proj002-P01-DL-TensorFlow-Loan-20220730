```python
def fill_mort_acc(total_acc, mort_acc):
    
    if np.isnan(mort_acc):
        return total_acc_avg[total_acc]#look up
    else:
        return mort_acc
```

```python
df_1['mort_acc'] = df_1.apply(lambda x: fill_mort_acc(x['total_acc'],x['mort_acc']),axis=1)
```

```python
X_train,X_test,y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

## Normalizing the Data - Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

## TensorFlow - Keras -Training Model

```python
pip show shap
```

```python
# models
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# permutation
#from sklearn.inspection import permutation_importance

# shap
import shap


# scoring metric
from sklearn.metrics import mean_squared_error
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(76, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(38, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(19, activation='relu'))
model.add(Dropout(0.2))

#BINARY CLASSIFICATION
model.add(Dense(1,activation = 'sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy') # as our y or label is continous-regression we use mean square error  
```

#### Training Model

```python
model.fit(x = X_train, y = y_train, 
          validation_data = (X_test,y_test),
          epochs=25, batch_size=256) #epochs in batch for large data
```
